```python
from numpy import genfromtxt
import numpy as np
mydata = genfromtxt('MoGCNcopy/result/SNF_fused_matrix.csv', delimiter=',')
print(mydata)
print(type(mydata))
```

```python
adjacency = mydata[1:,1:]
print(adjacency)
```

```python
import matplotlib.pyplot as plt
import networkx as nx

def show_graph(adjacency_matrix):
    rows, cols = np.where(adjacency_matrix >= 0.00001)
    edges = zip(rows.tolist(), cols.tolist())
    gr = nx.Graph()
    gr.add_edges_from(edges)
    nx.draw(gr, node_size=100, with_labels=False)
    plt.show()
```

```python

show_graph(adjacency)
```
